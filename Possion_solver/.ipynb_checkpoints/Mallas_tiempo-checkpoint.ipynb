```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Organización de los datos
mallas = np.array([400, 700, 1000])

# Diccionario de tiempos (segundos)
datos_ejercicios = {
    "Ejercicio 1": [16.3305, 42.2586, 86.4030],
    "Ejercicio 2": [5.6702, 14.6251, 35.9491],
    "Ejercicio 3": [5.2973, 13.7404, 32.2162],
    "Ejercicio 4": [5.8790, 18.3472, 36.4926]
}

# 2. Configuración estética de la gráfica
plt.figure(figsize=(10, 7), dpi=100)
plt.style.use('seaborn-v0_8-whitegrid') # Estilo limpio

colores = ['#d62728', '#1f77b4', '#2ca02c', '#9467bd'] # Rojo para el 1 (destaca), azul, verde, morado
marcadores = ['o', 's', '^', 'D']

# 3. Graficación
for i, (nombre, tiempos) in enumerate(datos_ejercicios.items()):
    plt.plot(mallas, tiempos, 
             label=nombre, 
             color=colores[i], 
             marker=marcadores[i], 
             linestyle='-', 
             linewidth=2, 
             markersize=8,
             alpha=0.8)

    # Añadir los valores de tiempo sobre cada punto para evitar ambigüedades
    for x, y in zip(mallas, tiempos):
        plt.text(x, y + 2, f'{y:.2f}s', ha='center', va='bottom', fontsize=9, fontweight='bold')

# 4. Personalización de ejes y títulos
plt.title('Rendimiento Computacional: Tiempo vs. Tamaño de Malla', fontsize=14, pad=20)
plt.xlabel('Dimensión de la Malla ($N$ en $N \\times N$)', fontsize=12)
plt.ylabel('Tiempo de Ejecución (segundos)', fontsize=12)

plt.xticks(mallas) # Asegura que solo se marquen los tamaños probados
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(frameon=True, facecolor='white', edgecolor='black')

# 5. Guardar y mostrar
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Datos (Sumatoria de mallas 400 + 700 + 1000)
datos_ejercicios = {
    "Ejercicio 1": [16.3305, 42.2586, 86.4030],
    "Ejercicio 2": [5.6702, 14.6251, 35.9491],
    "Ejercicio 3": [5.2973, 13.7404, 32.2162],
    "Ejercicio 4": [5.8790, 18.3472, 36.4926]
}

ejercicios = list(datos_ejercicios.keys())
tiempos_totales = [sum(t) for t in datos_ejercicios.values()]

# 2. Configuración estética
plt.figure(figsize=(9, 6), dpi=100)
plt.style.use('seaborn-v0_8-whitegrid')
colores = ['#d62728', '#1f77b4', '#2ca02c', '#9467bd']

# 3. Graficación de barras
barras = plt.bar(ejercicios, tiempos_totales, color=colores, alpha=0.85, edgecolor='black')

# Añadir etiquetas de tiempo total sobre cada barra
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 2, 
             f'{yval:.2f}s', ha='center', va='bottom', fontweight='bold', fontsize=10)

# 4. Personalización
plt.title('Costo Computacional Total (Suma de todas las mallas)', fontsize=14, pad=20)
plt.ylabel('Tiempo Total (segundos)', fontsize=12)
plt.xlabel('Ejercicios', fontsize=12)
plt.ylim(0, max(tiempos_totales) * 1.15) # Espacio para las etiquetas
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
